```python
import matplotlib.pyplot as plt
import numpy as np


x = np.linspace(0, 2 * np.pi, 100)
y1, y2 = np.sin(x), np.cos(x)


fig, ax = plt.subplots()
ax.plot(x, y1)
ax.plot(x, y2)
plt.show()
```

```python
x = np.linspace(0, 2 * np.pi, 100)
y1, y2 = np.sin(x), np.cos(x)

fig, (ax1 ,ax2) = plt.subplots(2, 1, figsize=(7.5, 7))
ax1.plot(x, y1, color="C0", linestyle="--")

ax2.plot(x, y2, color="C1")

ax1.set_ylabel("seno")
ax1.set_xlabel("Radio")

plt.show()
```

```python
x = np.linspace(0, 2 * np.pi, 100)
y1, y2 = np.sin(x), np.cos(x)

fig, (ax1 ,ax2) = plt.subplots(1, 2)
ax1.plot(x, y1, color="C0", linestyle="--", linewidth=1, marker="o")
ax2.plot(y2, x, color="C1")

plt.show()
```

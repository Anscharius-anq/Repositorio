### Figures 
 
 El objeto *Figure* es el contenedor principal que actúa como el lienzo donde se organizan todos los elementos de la visualización (*es como pensar en él como la hoja de papel donde se dibuja los gráficos.*).
 `Figure` puede contener uno o más objetos `Axes` , que son las áreas individuales donde realmente se grafican los datos. Cada Axes tiene sus propios ejes (x e y), etiquetas, títulos y los datos que quieres mostrar.

### Axes

Cada `Axes` es una región específica dentro de la `Figure` donde se dibuja un gráfico individual. Es aquí donde configuras detalles como los límites de los ejes (`set_xlim()`, `set_ylim()`), añades etiquetas (`set_xlabel()`, `set_ylabel()`) o defines el título del gráfico (`set_title()`). Los datos se agregan al Axes usando métodos como `plot()` para líneas, `scatter()` para puntos o `bar()` para gráficos de barras. También puedes añadir anotaciones o texto directamente en el Axes para resaltar información importante.

## Crear Figures y Axes:

`fig, ax = plt.subplots()` se crea una Figure (el lienzo) y por defecto se genera un solo Axes (el área de gráficos)

```python
fig, ax = plt.subplots()
```

### se puede especificar el números de Axes:

`fig, ax = plt.subplots(ncolumnn narrow)`

```python
fig, ax = plt.subplots(3, 2)
```

## Editar Spines

```python
fig, ax = plt.subplots()
ax.spines["top"].set_color("None")
ax.spines["right"].set_color("None")
```

```python

fig, axs = plt.subplots(3, 3)
axs[1, 2].plot([1, 2, 3])
gs = axs[1, 2].get_gridspec()
# remove the underlying Axes
for ax in axs[1:, -1]:
    ax.remove()
axbig = fig.add_subplot(gs[1:, -1])
axbig.annotate('Big Axes \nGridSpec[1:, -1]', (0.1, 0.5),
               xycoords='axes fraction', va='center')

fig.tight_layout()
```

```python
fig , axs = plt.subplots(3, 3, figsize=(8, 7))

for row in range(1, 3):  # Filas desde la 1 hasta la 2 (índices 1 y 2)
    for col in range(1, 3):  # Columnas desde la 1 hasta la 2 (índices 1 y 2)
        axs[row, col].remove()

gs = axs[1, 1].get_gridspec()
axbig = fig.add_subplot(gs[1:, 1:])

```
